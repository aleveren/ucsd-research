```python
import sympy as sp
```

```python
def log_godel_to_seq(g):
    ints = np.exp(np.atleast_1d(g)).round().astype('int')
    result = []
    for i in ints:
        f = sorted(sp.factorint(i).items(), key=lambda x: x[0])
        result.append(np.array([fi[1] for fi in f]) - 1)
    return result
```

```python
for a, b in zip(log_godel_to_seq(res[2]), log_godel_to_seq(res[3])):
    assert np.array_equal(b[:-1], a)
    print a, b
```
