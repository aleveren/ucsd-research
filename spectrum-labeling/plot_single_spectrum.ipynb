```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pandas as pd
%matplotlib notebook
```

```python
from data_util import row_to_spectrum
```

```python
save_figures = True
```

```python
x0, x1 = 420, 440
y0, y1 = -0.2e12, 2.5e12
```

```python
row = pd.read_csv("../data/CCS/subsetShots_5pct.csv", nrows=100).iloc[0, :]
x, y = row_to_spectrum(row, gap_insert_nan=2)

print np.where(np.isnan(x))[0]
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(12, 6, forward = True)

ax.add_patch(patches.Rectangle(
    (x0, y0), x1 - x0, y1 - y0,
    alpha = 0.5,
    edgecolor = "none",
    facecolor = "gray",
))

ax.plot(x, y, 'b-')
#ax.set_yscale('log')
ax.set_xlim(200, 950)
ax.set_ylim(-0.05e13, 1.5e13)
ax.set_xlabel("Wavelength (nm)")
ax.axhline(y=0, linestyle=':', color='k')
ax.set_title("Example spectrum (detailed region shown in gray)")

if save_figures:
    plt.savefig("example_spectrum.png")

None
```
