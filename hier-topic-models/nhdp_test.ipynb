```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import nhdp
```

```python
np.random.seed(1)
```

```python
vocab = ["is", "the", "a", "dog", "cat", "fast", "slow", "large", "small"]
```

```python
nhdp_obj = nhdp.NHDP(
    alpha = 5.0,
    beta = 5.0,
    gamma1 = 2.0,
    gamma2 = 1.0,
    topic_alpha_vector = 0.001 * np.ones(len(vocab)),
    vocab = vocab)
```

```python
#nhdp_obj.draw_document(document_length=10)
#nhdp_obj.draw_corpus(num_documents=1, document_length=1)
nhdp_obj.draw_corpus(num_documents=100, document_length=10)
```

```python
from scipy.stats import beta, dirichlet
```

```python
x = np.linspace(0,1,1000)
y = beta.pdf(x, a = nhdp_obj.gamma1, b = nhdp_obj.gamma2)
plt.plot(x,y)
plt.gca().set_title("Density function of Beta({:g}, {:g})".format(nhdp_obj.gamma1, nhdp_obj.gamma2));
```
